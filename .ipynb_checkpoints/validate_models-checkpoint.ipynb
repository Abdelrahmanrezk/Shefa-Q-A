```python
passage_question_objects = read_JSONL_file('datasets/qrcd_v1.1_dev.jsonl')
```

```python
df_file = json_to_dataframe(passage_question_objects)
```

```python
df_file.head()
```

```python
dataset = Dataset.from_pandas(df_file)
```

```python
dataset[0]
```

```python
model_checkpoint = "models/best_res_model_24_3_torch"
# tko = "bhavikardeshna/multilingual-bert-base-cased-arabic"

tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = AutoModelForQuestionAnswering.from_pretrained(model_checkpoint)
pipe = pipeline("question-answering", model=model, tokenizer=tokenizer)
```

```python
results = {}
for obj in dataset:
    context   = obj['context']
    question  = obj['question']
    
    result = pipe(question=question, context=context, top_k=5)
    
    res = []
    for i, ans in enumerate(result):
        ans.pop("start")
        ans.pop("end")
        ans["rank"] = (i+1)
        res.append(ans)
    results[obj["id"]] = res
```

```python
def dump_jsonl(data, output_path, append=False):
    """
    Write list of objects to a JSON lines file.
    """
    mode = 'a+' if append else 'w'
    with open(output_path, mode, encoding='utf-8') as f:
        for line in data:
            json_record = json.dumps(line, ensure_ascii=False)
            f.write(json_record + '\n')
    print('Wrote {} records to {}'.format(len(data), output_path))
```

```python
app_json = json.dumps(results, ensure_ascii=False, sort_keys=True)
print(app_json)
```
